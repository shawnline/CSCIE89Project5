```python
import gensim
import nltk

smallTextBytes: bytes = open("TeachingComputerScienceBrief", "rb").read()
largeTextBytes: bytes = open("TeachingComputerScienceFull", "rb").read()
# Remove non ascii characters
smallText: str = smallTextBytes.decode("ascii", "replace").lower()
largeText: str = largeTextBytes.decode("ascii", "replace").lower()
# Re encode as ASCII so each character maps to one byte
smallTextCharValues: list[int] = list(smallText.encode("ascii", "replace"))
largeTextCharValues: list[int] = list(largeText.encode("ascii", "replace"))

```

```python
import numpy as np
seq_length: int = 10
X: list = []
y: list = []
for i in range(0, len(largeTextCharValues) - seq_length):
    X.append(largeTextCharValues[i:i + seq_length])
    y.append(largeTextCharValues[i + seq_length])
```

```python
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense

# Define vocabulary size and embedding size
embedding_size = 8  # You can adjust this value

# Build the model
model = Sequential()
model.add(Embedding(input_dim=255, output_dim=embedding_size))
model.add(LSTM(100))
model.add(Dense(255, activation='softmax'))

# Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(np.array(X), np.array(y), epochs=20, batch_size=128)
```

```python
embeddings = model.layers[0].get_weights()[0]
print(embeddings[:30])
```

```python
# Initialize a list to hold embeddings for the SmallArticle
small_text_embeddings = []

for idx in smallTextCharValues:
    if idx != -1:  # If the character is known
        embedding_vector = embeddings[idx]
    else:
        embedding_vector = np.zeros(embedding_size)
    small_text_embeddings.append(embedding_vector)
# Convert to numpy array
small_text_embeddings = np.array(small_text_embeddings)
```
